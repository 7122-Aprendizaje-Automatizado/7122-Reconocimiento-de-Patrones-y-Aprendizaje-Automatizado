# Vidrios magnéticos y Boltzmann

## ¿Qué es la mecánica estadística?
La mecánica estadística es la subdisciplina de la física que se
estudia como la descripción de los fenómenos macroscópicos se
vincula con las leyes de la mecánica que gobiernan la dinámica
de los constituyentes (partículas, átomos o moléculas).

La mecánica estadística se inició en 1870 con los trabajos del físico austriaco Ludwig Boltzmann, muchos de los cuales fueron publicados en conjunto en el 1896. El término “termodinámica estadística” se propuso para uso por J. Willard Gibbs en 1902.


Desde la perspectiva microscópica se tienen las variables $x$, $v$

Desde la perspectiva macroscópica se tienen las variables $T$, $P$, $V$, $m$, etc.

### Vidrios magnéticos

## Interacción efectiva Ruderman-Kittel-Kasuya-Yosida-Van Vieck (RRKY)

$$J(r_{ij})=J_o\frac{cos(2k_FJ_{ij}+\phi_0)}{(k_FJ_{ij})^3}$$

Tenemos $N$ impurezas magnéticas colocadas al azar en
el material no magnético, y cada una puede tomar solo
dos valores
$$S_i(t) = \pm 1$$
El estado del material a tiempo $t$ se representa por un
vector de $N$ componentes $\pm 1$

$$\bar{S}(t) = (S_1(t), S_2(t), \cdots, S_N(t))$$

Los acoplamientos de intercambio RRKY se eligen
al azar con distribución gaussiana con esperanza $E[X]=J_0$ y varianza $Var[X]=J^2$
$$P(J_{ij})=\sqrt(\frac{N}{2\pi J^2})exp(-\frac{N}{2J^2}(J_{ij}-\frac{J_0}{N})^2)$$

La energía entre dos impurezas magnéticas está dada por su interacción antiferromagnética entre ambos spines

$$\epsilon_{ij} = -J_{ij}S_iS_j$$

La energía total del sistema de $N$ impurezas, mejor conocido como el hamiltoniano $H$ es
$$H=\sum_{ij}^N\epsilon_{ij} = -\sum_{ij}^N J_{ij}S_iS_j$$

La función de partición $Z$ describe las propiedades estadísticas de un sistema en equilibrio termodinámico. Un sistema termodinámicamente grande está en contacto térmico con el medio ambiente, con una temperatura $T$, y tanto el volumen del sistema como el número de partículas constituyentes son fijos. Una colección de este tipo de sistema comprende un conjunto denominado conjunto canónico. Para los microestados $\alpha$ del sistema, la función de partición es:
$$Z=\sum_\alpha exp(-\beta H_\alpha),$$
donde $\beta=\frac{1}{k_BT}$

Además, la probabilidad de cada microestado con la energía $H_{\alpha}$ es
$$P_{\alpha} = \frac{1}{Z}exp(-\frac{H_{\alpha}}{k_BT})$$

Por tanto, el valor medio asociado a esta distribución de energía es:
$$<H> = \sum_{\alpha}H_{\alpha}P_{\alpha} = -\frac{1}{Z}\frac{\partial Z}{\partial \beta} = -\frac{\partial lnZ}{\partial\beta}=k_BT^2\frac{\partial lnZ}{\partial Z}$$

La varianza es 

$$Var[H] = <(H-<H>)^2>=\frac{\partial^2 lnZ}{\partial\beta^2}$$

### Entropía
La entropía $S$ mide el grado en que la probabilidad del sistema se distribuye entre los diferentes microestados posibles. En contraste con el macroestado, que caracteriza cantidades promedio claramente observables, un microestado especifica todos los detalles moleculares sobre el sistema, incluidas la posición y la velocidad de cada molécula. Cuantos más estados de este tipo estén disponibles para el sistema con una probabilidad apreciable, mayor será la entropía. En mecánica estadística, la entropía es una medida del número de formas en que se puede organizar un sistema, a menudo tomada como una medida de "desorden" (cuanto mayor es la entropía, mayor es el desorden):

$$S=-k_B\sum_{\alpha}P_{\alpha}ln P_{\alpha}=k_B(lnZ+\beta <H>)=k_B\frac{\partial TlnZ}{\partial \beta}$$

## Energía libre de Helmholtz
El cambio en la energía de Helmholtz durante un proceso es igual a la cantidad máxima de trabajo que el sistema puede realizar en un proceso termodinámico en el que la temperatura se mantiene constante. A temperatura constante, la energía libre de Helmholtz se minimiza en el equilibrio.

$$F = <H>-TS = -k_BTlnZ$$

### Regresando al problema inicial...

A temperatura cero, el spin de cada impureza se alinea con el campo local que siente:

$$S_i = signo(h_i),$$

donde 

$$h_i = \sum_{j}^NJ_{ij}S_j + b_i$$

A temperatura distinta de cero, cada spin tiene una dinámica estocástica que depende
del campo local que siente:

$$P(S_i = \pm 1) = \frac{1}{1+exp(\mp\frac{2h_i}{k_BT})}$$
