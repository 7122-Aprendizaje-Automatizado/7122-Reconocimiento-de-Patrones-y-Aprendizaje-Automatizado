# Sistemas dinámicos

Es un sistema cuyo estado evoluciona en el tiempo. No el símil de la Física (derivado de fuerzas), sino cambios temporales.

Lo interesante es analizar cómo evolucionan en el tiempo las neuronas, sus componentes, sus señales.

## Sistemas dinámicos continuos

Son descritos por funciones continuas y por sus razones de cambio. El tiempo es la variable independiente. 

Las variables dependen de las mismas variables a buscar.

$$\frac{dx(t)}{dt}=ax(t)(1-x(t))$$

## Sistemas dinámicos discretos

Son descritos por relaciones sucesivas. El tiempo, variable independiente, es discreto.

$$x(t+1)=ax(t)(1-x(t))$$


La forma general de la ecuación diferencial ordinaria es
$$\frac{dx(t)}{dt}=f(x(t),t)$$

El tiempo aparece explícitamente.

Si el tiempo no aparece explícitamente, se dice que el sistema dinámico y la ecuación son AUTÓNOMOS.

## Sistema de n ecuaciones diferenciales ordinarias acopladas (EDO)
$$\frac{dx_1(t)}{dt}=f(x_1(t),x_2(t)\cdots , x_n(t))$$
$$\frac{dx_2(t)}{dt}=f(x_1(t),x_2(t)\cdots , x_n(t))$$
$$\cdots$$
$$\frac{dx_n(t)}{dt}=f(x_1(t),x_2(t)\cdots , x_n(t))$$

Existen casos donde se involucran derivadas de mayor orden

$$\frac{d^nx(t)}{dt^n}=f(x(t),\frac{dx(t)}{dt},\cdots , \frac{d^nx(t)}{dt^n})$$

## Ejemplos

En física, la dinámica newtoniana está descrita por EDO de segundo orden:

$$m\frac{d^2\bar{x}}{dt^2}=\bar{F}(\bar{x}(t),t),$$

donde

$$\bar{x}(t) = (x_1(t),x_2(t),x_3(t)),$$ y
$$\bar{F}(t) = (F_1(t),F_2(t),F_3(t))$$


## Notación

$$\frac{dx}{dt} = \dot{x}$$
$$\frac{d^2x}{dt^2} = \ddot{x}$$

## Puntos importantes

No es de nuestro interés encontrar soluciones analíticas de los sistemas EDO.

En su lugar, es de interés saber dónde estará el sistema cuando $t\rightarrow\infty$ y conocer la naturaleza dinámica de los atractores. Si $f(x)$ es lineal, el problema es algebraicamente sencillo de resolver; sin embargo, la mayoría de los problemas consideran funciones no lineales.

## Ejemplo

$$\ddot{x}=-kx$$
$$\dot{x}=\pi$$

## Dimensión

La dimensión del sistema dinámico se refiere a la dimensión del espacio en la cual se debe representar el problema, es decir, el número de variables dinámicas. NO se refiere a la dimensión espacial de representación.

Por ejemplo, una neurona con 30 entradas vive en $R^{30}$ desde el punto de vista dinámico.

Por tanto, nos interesa cómo la **no linealidad** y la **dimensión** afectan la naturaleza del comportamiento de un sistema a lo largo del tiempo.

Buscamos trayectorias particulares que describan el cambio $x(t)$, para lo cual haremos uso de métodos númericos con el problema de valor inicial que resuelvan este problema, como son **Euler** y **Runge Kutta**

## Problema del valor inicial

Considerando $\dot{x}=f(x)$, debemos encontrar o aproximar la trayectoria que vale $x_0$ en $t_0$ en el intervalo $t_0\in[t_i,t_f]$.

## Definición formal

Un sistema dinámico es el espacio de estados $X$, un conjunto de los tiempos $T$ y una regla $f$ que especifica cómo el espacio de los estados $X$ evoluciona en el tiempo. La regla $f$ es una función con dominio $X\times T$ e imagen $X$:
$$f:X\times T \rightarrow X$$

# Atractores, repulsores y puntos fijos

Consideremos un ejemplo:

$$\dot{x}= \sin(x)$$

Si $f(x*)=0$, entonces $x*$ es un punto fijo del sistema dinámico.

Los puntos 'llenos' son **ATRACTORES (estables)**.


Los puntos 'vacíos' son **REPULSORES (inestables)**.

El **FLUJO** se define en función del signo de $f(x)$.

## ¿Qué sucede en $x=\frac{\pi}{4}$?



El sistema de acerca asintóticamente a $\pi$, pero nunca llega, pues a medida que se aproxima a este punto, la velocidad tiende a cero. De esta forma podemos describir todas las trayectorias cuando $t\rightarrow\infty$.

## Atractores 

En este caso son 
$$x*_K = k\pi,$$
donde $$k\in Z$$

Los atractores estables atraen el flujo de infinitas trayectorias.

Por lo tanto, podemos resumir que, cuando $t\rightarrow\infty$ el sistema puede

1. Ir hacia infinito
2. Ir hacia menos infinito
3. Ir o permanecer en un punto fijo estable
4. Permanecer en un punto fijo inestable

# Campos vectoriales

A cada punto $x$ se le asigna un vector con magnitud $|f(x)|$ y con dirección correspondiente a la razón de cambio de $f(x)$. Estos vectores nos definen el campo vectorial del sistema dinámico.

# Análisis de estabilidad lineal



$$x(t) = x* +\eta(t)$$

$$\dot{x}=f(x)$$
$$\dot{(x* +\eta(t))} = f(x* +\eta(t))$$
$$\dot{\eta(t)} = f(x* +\eta(t))$$

Si tomamos $\eta(x)<<1$, entonces usamos el teorema de Taylor para aproximar la función por una recta. Además, se necesita que $f(x)$ sea continua, diferenciable y también su derivada y su segunda derivada.


$$f(x* +\eta(t))=f(x*)+\eta(x)f'(x*)+\frac{1}{2}\eta^2(x)f''(x)+O(3)$$

Tomando en cuenta que $\eta(x)<<1$, despreciamos el error y entonces

$$f(x*+\eta(x))=\eta(x)f'(x)$$

Por tanto, la **ecuación de la perturbación** es

$$\dot{\eta(x)}\approx \eta(x)f'(x*)$$

La solución a esta ecuación es

$$\eta(x) = Ae^{f'(x*)t}$$

Si $f'(x*)>0$, la perturbación crece exponencialmente.

Si $f'(x*)<0$, la perturbación decrece exponencialmente.

Además. $\frac{1}{|f'(x*)|}$ es el tiempo característico.
