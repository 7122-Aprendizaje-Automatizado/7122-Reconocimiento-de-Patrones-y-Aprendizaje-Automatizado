```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
from imblearn.over_sampling import SMOTE, ADASYN
```

```python
X, y = make_classification(n_samples=500, n_features=2, n_informative=2,
                           n_redundant=0, n_repeated=0, n_classes=2,
                           n_clusters_per_class=1,
                           weights=[0.20, 0.80],
                           #weights=[0.20, 0.30, 0.50],
                           class_sep=1.5, random_state=0)
plt.scatter(X[:,0],X[:,1],c=y,alpha=.7)
print(sorted(Counter(y).items()))
```

#SobreMuestros

```python
#Sobre Muetreo "Ingenuo"
ros = RandomOverSampler(random_state=0)
X_resampled, y_resampled = ros.fit_resample(X, y)
plt.scatter(X_resampled[:,0],X_resampled[:,1],c=y_resampled,alpha=.4)
#Contamos las clases y cuantos hay
print(sorted(Counter(y_resampled).items()))
```

```python
#Uso de SMOTE
X_resampled, y_resampled = SMOTE().fit_resample(X, y)
print(sorted(Counter(y_resampled).items()))
plt.scatter(X_resampled[:,0],X_resampled[:,1],c=y_resampled,alpha=.85)
```

```python
#Uso de ADASYN
X_resampled, y_resampled = ADASYN().fit_resample(X, y)
print(sorted(Counter(y_resampled).items()))
plt.scatter(X_resampled[:,0],X_resampled[:,1],c=y_resampled,alpha=.85)
```

#SubMuestro

```python
#RandomUnderSampling
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=0)
X_resampled, y_resampled = rus.fit_resample(X, y)
plt.scatter(X_resampled[:,0],X_resampled[:,1],c=y_resampled,alpha=.7)
# Contamos las clases y cuantos hay.
print(sorted(Counter(y_resampled).items()))
plt.show()
```

```python
from imblearn.under_sampling import CondensedNearestNeighbour
cnn = CondensedNearestNeighbour(sampling_strategy='auto',n_neighbors=5)
X_resampled, y_resampled = cnn.fit_resample(X, y)

plt.scatter(X_resampled[:,0],X_resampled[:,1],c=y_resampled,alpha=.7)
# Contamos las clases y cuantos hay.
print(sorted(Counter(y_resampled).items()))
plt.show()
```

```python
from imblearn.under_sampling import TomekLinks
tl = TomekLinks(sampling_strategy='all')
X_resampled, y_resampled = tl.fit_resample(X, y)

plt.scatter(X_resampled[:,0],X_resampled[:,1],c=y_resampled,alpha=.7)
# Contamos las clases y cuantos hay.
print(sorted(Counter(y_resampled).items()))
plt.show()
```

# SMOTE-TOMEK

```python
#SMOTE-TOMEK
X_resampled, y_resampled = SMOTE().fit_resample(X, y)
X_resampled, y_resampled = TomekLinks(sampling_strategy='all').fit_resample(X_resampled, y_resampled)

plt.scatter(X_resampled[:,0],X_resampled[:,1],c=y_resampled,alpha=.7)
# Contamos las clases y cuantos hay.
print(sorted(Counter(y_resampled).items()))
plt.show()
```
