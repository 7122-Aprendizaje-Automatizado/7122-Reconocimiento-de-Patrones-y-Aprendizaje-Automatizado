#Conjuntos de punto flotante

Los números de punto flotante se representan en el ordenador usando un número finito de bits, lo que significa que hay un número limitado de valores que se pueden representar de manera precisa. Este conjunto de números se conoce como el conjunto de punto flotante y se define en el estándar IEEE 754.


```python
import math

a = 0
b = 1.e-400

if math.isclose(a, b):
    print(f"Los números son iguales.{(a,b)} -- {a==b}")
else:
    print(f"Los números no son iguales.{(a,b)} -- {a==b}")
```

Es importante tener en cuenta que los números de punto flotante no son exactos, y debemos ser conscientes de esto al trabajar con ellos en nuestro código.


#Polinomio de Taylor

-----------------------------------------------------------------------------------------------------------

Este bloque utiliza la serie de Taylor/Mac Laurin para $sen(x)$ para calcular el valor de la función para un valor dado de x y un número dado de términos de la serie. Se compara el resultado con el valor real de $sen(x)$ utilizando la función math.isclose(). Si ejecutas el bloque, verás que el resultado es cercano pero no idéntico al valor real de la función debido a la imprecisión de los números de punto flotante.

$sen(x) = Σ_{i=0}^∞ \frac{(-1)^i x^{2i+1}}{(2i+1)!} $

$cos(x) = Σ_{i=0}^∞ \frac{(-1)^i x^{2i}}{(2i)!} $


```python
def taylor_sin(x, n):
    result = 0
    for i in range(n):
        term = ((-1)**i)*x**(2*i+1)/(math.factorial(2*i+1))
        result += term
    return result

x = math.pi/4
grado = 6
print("Valor de la función sin({}) usando la serie de Taylor:".format(x))
print(taylor_sin(x, grado))
print("Valor real de la función sin({}):".format(x))
print(math.sin(x))
```

#  Newton-Raphson para raices de forma aproximada e iterativa.

En este ejemplo, la función f(x) representa la ecuación $sen(x) = 0$ que queremos resolver, y la función df(x) representa su derivada, $cos(x)$. La función newton_raphson implementa el método de Newton-Raphson para encontrar una solución de la ecuación f(x) = 0, empezando desde un punto inicial x0. La función devuelve None si no se pudo encontrar una solución dentro del número máximo de iteraciones especificado.

Si ejecutas el programa, verás que la aproximación de la solución de $sen(x) = 0$ es cercana a pi, pero no idéntica debido a la imprecisión de los números de punto flotante.

```python
def f(x):
    return math.sin(x) # np.cos(x)

def df(x):
    return math.cos(x) # -np.sin(x)

def newton_raphson(f, df, x0, tol, max_iter):
    x = x0
    for i in range(max_iter):
        fx = f(x)
        if abs(fx) < tol:
            return x
        dfx = df(x)
        if dfx == 0:
            break
        x = x - fx/dfx
    return None

x0 = 4 # x_0 que hay que especificar.
tol = 1e-8
max_iter = 1000

x = newton_raphson(f, df, x0, tol, max_iter)

if x is not None:
    print("Aproximación de la solución de sin(x) = 0:")
    print(x)
    print("Valor real de la solución de sin(x) = 0:")
    print(math.pi)
    if math.isclose(x,math.pi):
      print("Los valores son cercanos.")
    else:
      print("Los valores no son cercanos.")
else:
    print("No se pudo encontrar una solución.")
```
